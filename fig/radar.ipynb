```python
import matplotlib.pyplot as plt
import numpy as np

# 
categories = [
    "en-ar", "en-de", "en-fr", "en-it", "en-zh", 
    "ar-en", "de-en", "fr-en", "it-en", "zh-en"
]

values1 = [4.5, 15.6, 35.1, 26.2, 33.2, 28.2, 29.6, 36.2, 30.1, 34.5]
values2 = [3.5, 13.4, 33.2, 22.5, 30.2, 27.2, 29.0, 34.1, 29.5, 33.2]

# 
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))

# 
theta = np.linspace(0, 2*np.pi, len(categories), endpoint=False)

# 
values1 = np.concatenate((values1, [values1[0]]))
values2 = np.concatenate((values2, [values2[0]]))
theta = np.concatenate((theta, [theta[0]]))

# 
ax.plot(theta, values1, marker='o', linestyle='-', color='#1f77b4', linewidth=2, label='LANDeRMT')
ax.plot(theta, values2, marker='s', linestyle='--', color='#ff7f0e', linewidth=2, label='Adapter-LoRA')

# 
ax.fill(theta, values1, alpha=0.2, color='#1f77b4')
ax.fill(theta, values2, alpha=0.2, color='#ff7f0e')

# 
ax.set_xticks(theta[:-1])
ax.set_xticklabels(categories, fontsize=10)  # 

# 
plt.legend(loc='upper right', bbox_to_anchor=(1.1, 1))
plt.savefig('radar.pdf', format='pdf', dpi=300)

# 
plt.show()
```
